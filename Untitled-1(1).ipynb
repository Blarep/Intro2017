```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    s1 = np.cos(2*np.pi*t)
    e1 = np.exp(-t)
    return np.multiply(s1, e1)
```

```python
fig = plt.figure(figsize=(30,15))
fig.suptitle('Graficazos!', fontsize=112);
```

```python
ax = fig.add_subplot(1, 2, 1)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.0, 2.0, 0.01)

ax.plot(t1, f(t1), 'bo',
        t2, f(t2), 'k--', markerfacecolor='green')
ax.grid(True)
ax.set_ylabel('Damped oscillation');
```

```python
ax = fig.add_subplot(1, 2, 2, projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1, 1)

plt.show()
```

```python
from ipywidgets import interact
```

```python
Y = np.linspace(0,10,100)
def f(t,k1,k2,v1,v2):
    fig = plt.figure(figsize=(25,20))
    fig.suptitle('Interferencia de ondas', fontsize=60);
    plt.figure(figsize=(20,5))
    x = np.linspace(0,10,200)
    ax1 = fig.add_subplot(2, 2, 1)
    y1 = np.sin(k1*(x+v1*t))
    ax2 = fig.add_subplot(2, 2, 3)
    y2 = np.sin(k2*(x-v2*t))
    ax3 = fig.add_subplot(1, 2, 2)
    y3 = y1 + y2
    ax1.plot(x, y1, "r")
    ax2.plot(x, y2, "g")
    ax3.plot(x, y3, "b")
    ax1.set_ylim(-2, 2)
    ax2.set_ylim(-2, 2)
    ax3.set_ylim(-2, 2)
    ax1.set_ylabel('y(x,t) = sin(k(x+vt))', fontsize=35);
    ax2.set_ylabel('y(x,t) = sin(k(x-vt))', fontsize=35);
    ax3.set_ylabel('y(x,t) = sin(k(x+vt)) + sin(k(x-vt))', fontsize=35);
    ax1.set_xlabel('x', fontsize=35);
    ax2.set_xlabel('x', fontsize=35);
    ax3.set_xlabel('x', fontsize=35);
    plt.show()
interact(f, t=(0,10,0.1), k1=2, k2=2, v1=2*np.pi, v2=2*np.pi);
```
