## 1.- `¿Qué es Jupyter notebook?`

La `Jupyter notebook` es una aplicación web basada en una estructura de servidor-cliente. Su principal funcionalidad es permitirnos la creación y manipulación de documentos **notebook**. 

Un **notebook** es un documento que mezcla elementos de **texto**, código **HTML** y código ejecutable de __Julia__, __Python__ y __R__ (_puede soportar más lenguajes_), esto hace a `Jupyter notebook` el lugar ideal para _describir_ un análisis, _mostrar resultados_ y poder realizar el análisis en tiempo real.

<h4>¿Por qué usamos notebooks?</h4>

Los  `Jupyter notebooks` soportan editar texto ultizando el formato **markdown** ([más](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet)).

Ofrece soporte para $\LaTeX$, lo que nos permite escribir expresiones matemáticas como:

$$\text{Gradient Descent}\\\theta_j = \theta_j - \alpha \frac{\partial }{\partial \theta_j} J(\vec{\theta}) = \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m}(h_\theta(\vec{x^{(i)}}) - y^{(i)}) * x^{(i)}_{j}$$


Podemos agregar a nuestro documentos componentes **HTML**, lo que le permite agregar imágenes, videos o cualquer otro componente que soporte el lenguaje.

En cuanto a programación, no entrega una vía fácil para editar y ejecutar nuestro código gracias a su paradigma de **celdas** (_como Mathematica_).

````python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    num = n*factorial(n-1)
    print(num)
    return num

factorial(20);
````

Para la visualización de datos, módulos como **Matplotlib** y **Seaborn** (_entre muchas otras_) son soportadas por los `jupyter notebooks`, estos muestran sus gráficos dentro del documento.

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,4))

x = np.linspace(0, 16*np.pi, num=1000)
y1, y2 = np.sin(x), np.cos(x)

plt.plot(x,y1, "g")
plt.plot(x,y2, "r")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["sin(x)", "cos(x)"])

plt.show()
````

## 4.- `Usando notebooks`

Podemos ininiar el servidor de `Jupyter notebook` a través del icono de acceso directo para usuarios de `Windows`, o desde la terminal con `jupyter notebook` para usuarios de sistemas basados en `UNIX` o desde `Conda Promp`.

Esto nos lanzará al navegador donde podremos crear, abrir y borrar nuestros documentos.

[Jupyter Notebook Cheats Sheet](https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Jupyter_Notebook_Cheat_Sheet.pdf)

#### Modos

    - Modo Comando: Presionamos ESC
        Nos permite la edición de las celdas: Crear, eliminar, copiar, cambiar tipo, etc
    - Modo Edición: Presionamos ENTER
        Nos permite editar el contenido de la celda.
        
#### Celdas
    - Cada celda se ejecuta de manera independiente, ejecutar una celda no implica que todo el archivo se ejecute. 
    - Las variables se comparten entre celdas, pero si una variable es declarada en una celda no ejecutada, no puede usarse en otras celdas.
    - Para ejecutar una celda usamos [Ctrl + ENTER] o [SHIFT + ENTER].

#### Tipos de celdas

    - Markdown: Formato de texto enriquesido (HTML, LaTeX, MarkDown).
    - Code: Interprete de código.

## 6.- `Plots`

Como se dijo anteriormente, los `Jupyter notebooks` soportan los gráficos de **matplotlib**. Usamos el magic "%matplotlib inline" para mostrarlo en el mismo documento.

````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    s1 = np.cos(2*np.pi*t)
    e1 = np.exp(-t)
    return np.multiply(s1, e1)
````

````python
fig = plt.figure(figsize=(30,15))
fig.suptitle('Graficazos!', fontsize=80)

ax = fig.add_subplot(1, 2, 1)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.0, 2.0, 0.01)

ax.plot(t1, f(t1), 'bo',
        t2, f(t2), 'k--', markerfacecolor='green')
ax.grid(True)
ax.set_ylabel('Damped oscillation')
ax = fig.add_subplot(1, 2, 2, projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1, 1)

plt.show()
````

### `interact` IPython Widget

La ventaja de ser una aplicación web es la capacidad de ser intercativo. Podemos utilizar el submódulo  `ipywidgets.interact` para crear gráfico **interactivos**.

````python
from ipywidgets import interact
````

`interact` recibe como argumentos una función y los argumentos de esa función. Creará Widgets interactivos dependiendo el tipo de datos de los argumentos.

Comenzamos con un ejemplo simple de una función:

````python
def f(x):
    return x
````

**Int slice**: Dandole un valor `int`.

````python
interact(f, x=10);
````

**CheckBox**: Dandole un `bool`

````python
interact(f, x=True);
````

**Text Input**: Dandole un `string`

````python
interact(f, x="Hola");
````

**Dropdown List**: Dandole una lista o diccionario

````python
interact(f, x={"ESP":"Hola", "ENG":"Hello"});
````

También podemos usar varias variables y dar rangos.

````python
def f(w,x,y,z):
    return w,x,y,z

interact(f, w=(-10,10,1), x=(0,1,0.01),
            y="y", z=[1,2,3,4,5]);
````

Finalmente podemos combinar esto con nuestros conocimientos de **matplotlib** y crear gráficos interactivos.

````python
def plotter(t,k1,k2,v1,v2):
    fig = plt.figure(figsize=(20,10))
    fig.suptitle('Interferencia de ondas', fontsize=60);
    x = np.linspace(0,10,200)
    ax1 = fig.add_subplot(2, 2, 1)
    y1 = np.sin(k1*(x+v1*t))
    ax2 = fig.add_subplot(2, 2, 3)
    y2 = np.sin(k2*(x-v2*t))
    ax3 = fig.add_subplot(1, 2, 2)
    y3 = y1 + y2
    ax1.plot(x, y1, "r")
    ax2.plot(x, y2, "g")
    ax3.plot(x, y3, "b")
    ax1.set_ylim(-2, 2)
    ax2.set_ylim(-2, 2)
    ax3.set_ylim(-2, 2)
    ax1.set_ylabel('y(x,t) = sin(k(x+vt))', fontsize=20);
    ax2.set_ylabel('y(x,t) = sin(k(x-vt))', fontsize=20);
    ax3.set_ylabel('y(x,t) = sin(k(x+vt)) + sin(k(x-vt))', fontsize=20);
    ax1.set_xlabel('x', fontsize=20);
    ax2.set_xlabel('x', fontsize=20);
    ax3.set_xlabel('x', fontsize=20);
    plt.show()
    
    
interact(plotter, t=(0,10,0.01), k1=(1,5,0.01), k2=(1,5,0.01), v1=2*np.pi, v2=2*np.pi);
````
